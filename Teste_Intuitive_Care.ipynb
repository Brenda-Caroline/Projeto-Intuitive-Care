TESTE 1 - WEB SCRAPING

```python
#Instalando a biblioteca urllib3. Esta biblioteca possui recursos que nos permite conectar a um website e recuperar de lá o conteúdo de uma página.
!pip install urllib3
```

```python
#Importa a biblioteca
import urllib3
```

```python
#Armazena na variável url a url correspondente do site que quer baixar os arquivos
url = 'https://www.gov.br/ans/pt-br/assuntos/consumidor/o-que-o-seu-plano-de-saude-deve-cobrir-1/o-que-e-o-rol-de-procedimentos-e-evento-em-saude'
```

```python
#Agora estou criando a variável "conexao" e armazenando nela um objeto do tipo "PoolManager". Este objeto possui recursos para manipular conexões com websites.
conexao = urllib3.PoolManager()
```

```python
#Mais uma variável, porém esta agora guarda o conteúdo da página WEB. Observe que utilizamos uma funcionalidade chamada request que faz requisições ao site utilizando o método GET e informando a variável "url" criada anteriormente.
retorno = conexao.request('GET', url)
```

```python
#A variável retorno representa o resultado da conexão com a página. Utilizando o parâmetro "data", que o objeto de retorno possui, já daria para ter uma ideia do que foi recuperado. 
retorno.data
```

```python
#Instala biblioteca BeautifulSoup. Esta biblioteca nos facilita a manipulação de HTML da página onde desejamos baixar os arquivos.
!pip install BeautifulSoup
```

```python
#Nesta linha informamos que deve ser importado o recurso da biblioteca para manipular o código da página HTML.
from bs4 import BeautifulSoup
```

```python
#"pagina" será uma variável que armazenará o código HTML tratado pelo BeautifulSoup.
pagina = BeautifulSoup(retorno.data,"html.parser")
```

```python
pagina.prettify
```

```python
#Dado é uma lista vazia que será usada no algoritmo para guardar os hiperlinks recuperados.
#Esse array armazenará os links retornados que correspondem a tag "a" com classe = internal-link

dado = []
for link in pagina.find_all('a', class_ = 'internal-link'):
  dado.append(link.get('href'))
```

```python
#Retorna os hiperlinks que estão armazenados na lista
dado
```

```python
#Instala a biblioteca wget. Ela possui recursos para realizar download de dados de um endereço na internet.
!pip install wget
```

```python
import wget
```

```python
#Define o caminho que será armazenado os arquivos baixados. Nesse caso, foi criado uma pasta chamada "arquivos" no próprio ambiente do colab, onde ficam os arquivos e pastas utilizadas
caminho_destino = 'arquivos'
```

```python
#Deleta o links 1, 7 e 8 do array dado, pois esses links não nos interessa para baixar 
del dado[6:8]
del dado[0:1]
```

```python
#Retorna todo o conteúdo do array dado
dado
```

```python
#Percorre todo o array e faz o download de cada url na pasta definida anteriormente
for item in dado:
  wget.download(item, caminho_destino)
```
